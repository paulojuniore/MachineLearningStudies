```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Importamos uma ferramenta que cria conjuntos de dados aleátorios, separados por classes de centróides que são usados para aplicar o K Means Clustering algorithm para clasi

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8, random_state=50)
```

```python
plt.figure(figsize=(10,8))
sns.set_style("whitegrid")
plt.scatter(data[0][:, 0], data[0][:, 1], c=data[1])
```
