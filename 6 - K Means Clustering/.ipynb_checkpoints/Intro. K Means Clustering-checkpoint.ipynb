```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Importamos uma ferramenta que cria conjuntos de dados aleátorios, separados por classes de centróides que são usados para aplicar o K Means Clustering algorithm para classificação dos dados.

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8, random_state=50)
```

```python
data
```

### Abaixo podemos ver todos os dados separados por classe definida pela cor.

```python
plt.figure(figsize=(10,8))
sns.set_style("whitegrid")
plt.scatter(data[0][:, 0], data[0][:, 1], c=data[1])
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans =  KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(12,8))
ax1.set_title('K Means')
ax1.scatter(data[0][:, 0], data[0][:, 1], c=kmeans.labels_)
ax2.set_title('Original')
ax2.scatter(data[0][:, 0], data[0][:, 1], c=data[1])
```

### Após a utilização de algoritmo podemos ver que o resultado se aproximou muito do original.
