```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Utilizando dados de variáveis preditoras que influenciam no diagnóstico de pacientes com cancer, iremos utilizar support vector machines para a partir de novos valores o modelo possa predizer se um paciente futuro tem ou não cancer

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
cancer['feature_names']
```

```python
df_cancer = pd.DataFrame(data=cancer['data'], columns=cancer['feature_names'])
```

```python
df_cancer.info()
```

```python
df_cancer.head()
```

```python
df_target = pd.DataFrame(cancer['target'], columns=['Cancer'])
```

```python
df_target
```

```python
from sklearn.model_selection import train_test_split
```

## Dividindo os dados em dados de treino e teste, respectivamente, para execução do modelo
### np.ravel transforma o dataframe target em notação de array

```python
X_train, X_test, Y_train, Y_test = train_test_split(df_cancer, np.ravel(df_target), test_size=0.3, random_state=101)
```

```python
X_train
```

```python
Y_train
```

## Usando Support vector classifier para implementar o comportamento de SVM's no modelo em questão.
### gamma='scale' significa que o método usado para o kernel trick será a partir de uma heurística padrão.

```python
from sklearn.svm import SVC
```

```python
model = SVC(gamma='scale')
```

```python
model.fit(X_train, Y_train)
```

```python
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

### Precisão do modelo encontrada de 93%

```python
print(classification_report(Y_test, predictions))
print()
print(confusion_matrix(Y_test, predictions))
```

## Agora iremos utilizar o GridSearch, onde a partir da escolha de alguns valores de parâmetros do SVM, o algoritmo irá calcular para cada um dos valores da grade e assim, irá nos mostrar para quais valores de cada parâmetro serão mais adequados para uma melhoria na precisão do modelo.

### Abaixo temos os parâmetros para o teste de valores do GridSearch

```python
param_grid = {'C': [0.1, 1, 10, 100, 1000], 'gamma': [1, 0.1, 0.01, 0.001, 0.0001], 'kernel': ['rbf']}
```

```python
from sklearn.model_selection import GridSearchCV
```

### Criando a instância do GridSearch com os parâmetros informados anteriormente, o modelo será 'refitado' automaticamente para cada combinação e o 'verbose' significa que será mostrado um status da processo de análise das combinações de parâmetros.

```python
grid = GridSearchCV(SVC(), param_grid, refit=True, verbose=3)
```

```python
grid.fit(X_train, Y_train)
```

### Podemos ver abaixo os melhores valores para cada parâmetro após a classificação

```python
grid.best_params_
```

```python
pred = grid.predict(X_test)
```

### O modelo aumentou sua performance para 95% de precisão, antes de 93%.

```python
print(classification_report(Y_test, pred))
print()
print(confusion_matrix(Y_test, pred))
```
